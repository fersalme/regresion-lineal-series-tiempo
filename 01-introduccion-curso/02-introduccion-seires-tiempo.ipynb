```python
import pandas as pd
import matplotlib.pyplot as plt
import wbgapi as wb

data = wb.data.DataFrame(series= 'NY.GDP.MKTP.CD', economy = 'COL', time=range(1960, 2024), columns='series')
data = data.reset_index()
data = data.rename(columns={
    'time': 'Año',
    'NY.GDP.MKTP.CD': 'PIB (US$ a precios actuales)'
})

# Graficar la serie de tiempo
plt.figure(figsize=(11, 6))
plt.plot(data['Año'], data['PIB (US$ a precios actuales)'], marker='o')
plt.title('PIB de Colombia (1960-2024)')
plt.xlabel('Año')
plt.ylabel('PIB (US$ a precios actuales)')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```
